```python
# prompt: show 3 different catogories in a column chart comparing the percentage of growth for each seperately and then for all combined

import pandas as pd
import seaborn as sns

# Create the data frame
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'All'],
    'Growth': [0.1, 0.2, 0.3, 0.4]
})

# Create the column chart
sns.barplot(x='Category', y='Growth', data=df)

```
